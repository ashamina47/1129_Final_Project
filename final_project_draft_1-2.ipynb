```python
import turtle

def draw_rectangle(width, height, r,g,b, x = 0, y = 250):
    t = turtle.Turtle()
    t.speed(300)
    t.fillcolor(r,g,b)
    t.hideturtle()
    t.up()
    t.goto(x,y)
    t.down()
    t.begin_fill()
    for i in range(2):
        t.forward(width)
        t.right(90)
        t.forward(height)
        t.right(90)
    t.end_fill()
    t.setx(width)
    
def get_color(color):
    if color == "red":
        return 1,0,0
    if color == "white":
        return 1,1,1
    if color == "blue":
        return 0,0,1
    else:
        return 0,0,0
    
def draw_star(size, r,g,b, x = 0, y = 250):
    t = turtle.Turtle()
    t.speed(300)
    t.pencolor(get_color("white"))
    t.up()
    t.goto(x,y)
    t.hideturtle()
    t.down()
    t.begin_fill()
    t.fillcolor(r,g,b)
    for i in range(5):
        t.forward(size)
        t.right(144)     
    t.end_fill()

def draw_flag(height):
    flag_width = 1.9 * float(height)
    square_height = height * (7/13)
    square_width = flag_width * (2/5)
    stripe = height / 13
    star = stripe * (3/5)
    v = square_height / 9
    w = square_width / 14
    r,g,b = get_color("blue")
    draw_rectangle(square_width, square_height, r,g,b)
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, 250)
    r,g,b = get_color("white")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, 250-stripe)
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, 250-stripe*2)
    r,g,b = get_color("white")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, 250-stripe * 3)
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, 250-stripe * 4)
    r,g,b = get_color("white")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, 250-stripe * 5)
    r,g,b = get_color("red")
    draw_rectangle(flag_width - square_width, stripe, r,g,b, square_width, 250-stripe * 6)
    
    r,g,b = get_color("white")    
    draw_rectangle(flag_width, stripe, r,g,b, 0, 250-stripe*7)
    r,g,b = get_color("red")
    draw_rectangle(flag_width, stripe, r,g,b, 0, 250-stripe * 8)
    r,g,b = get_color("white")
    draw_rectangle(flag_width, stripe, r,g,b, 0, 250-stripe * 9)
    r,g,b = get_color("red")
    draw_rectangle(flag_width, stripe, r,g,b, 0, 250-stripe * 10)
    r,g,b = get_color("white")
    draw_rectangle(flag_width, stripe, r,g,b, 0, 250-stripe * 11)
    r,g,b = get_color("red")
    draw_rectangle(flag_width, stripe, r,g,b, 0, 250-stripe * 12)

    
    r,g,b = get_color("white")

    for i in range(1,13,2):
        for j in range(1,10,2):
            draw_star(star, r,g,b, v * i , 250-(w * j) )

    for i in range(2,12,2):
        for j in range(2,10,2):
            draw_star(star,r,g,b, v * i, 250-(w * j))

draw_flag(200)
```
